# College Majors Students Earnings - Data Visualization Project
## Description
We'll use a dataset on the job outcomes of students who graduated from college beteween 2010 and 2012. The original data on job outcomes was released by [American Community Survery](https://www.census.gov/programs-surveys/acs/), which conduct surveys and aggregates the data. 

The dataset used in this project can be find on this [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors) made by FiveThirtyEight, who cleaned the original dataset.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more.

The goal of this project is to use some visualization techniques to answer questions like:
- Do students in more popular majors make more money?
- How many majors are predominantly male? Predominantly female?
- Which category of majors have the most students?

We'll use a different type of plots for each of these questions. We'll primarily work with the *pandas* library, that has many methods for quickly generating common plots from data in DataFrames.

## Pandas Scatter Plots
We'll start by exploring some relations between columns by using the DataFrame.plot() method of the pandas library
### Sample_size vs. Median

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median')
```

### Sample_size vs. Unemployment_rate

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemployment_rate')
```

### Full_time vs. Median

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median')
```

### ShareWomen vs. Unemployment_rate

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment_rate')
```

### Men vs. Median

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median')
```

### Women vs. Median

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median')
```

### Total vs. Median

```python
recent_grads.plot(x='Total', y='Median', kind='scatter', title='Total vs. Median')
```

### ShareWomen vs. Median

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs. Median', xlim=(0,1))
```

### Comments

Based on the above plots, we can try to answer to some questions

**Do students in more popular majors make more money?**

As we can see from the plot *Full_time vs Median*, the answer to this question is no, since there is no positive correlation between the two.

**Do students that majored in subjects that were majority female make more money?**

As we can see from the last plot (*ShareWomen vs Median*), there is no evidence of this fact. Instead, it seems the opposite.

**Is there any link between the number of full-time employees and median salary?**

As we can see from the plot *Full_time vs. Median* there is no evidence of this fact.
